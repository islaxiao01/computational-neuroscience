```python
import numpy as np
import matplotlib.pyplot as plt
import random
random.seed(7967)
import matplotlib.pyplot as plt

data = np.loadtxt('Spike_raw_PCA_data.dat')

# check the number of rows (ndims) and columns (nvals):

ndims = np.size(data[0, :])

nvals = np.size(data[:, 0])
```

```python

with open('Spike_raw_PCA_data.dat', 'r') as file:
    data_lines = file.readlines()

data_lines[:10]

# Create a scatter plot of the two dimensions
plt.figure(figsize=(10, 8))
plt.scatter(data[:, 0], data[:, 1], alpha=0.6)
plt.title('2D Scatter Plot of PCA Simple Data')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.grid(True)
plt.show()
```

```python
C = np.cov(data, rowvar = False)

evals, evecs = np.linalg.eig(C)

C, evals, evecs
```

```python
# Sort the eigenvectors by their eigenvalues
idx = np.argsort(-abs(evals))

evals = evals[idx]

evecs = evecs[:, idx]
idx, evals, evecs
```

```python
# Display a graph of the eigenvalues, using plt.bar(x, y) to show a bar histogram.
pc_labels = [f'PC{i+1}' for i in range(len(evals))]

plt.figure(figsize=(8, 6))
plt.bar(pc_labels, evals, color=['blue'])
plt.title('Bar Histogram of Eigenvalues')
plt.xlabel('Principal Components')
plt.ylabel('Eigenvalue')
plt.grid(True, axis='y')
plt.show()

```

```python
# Transform the raw data values by multiplying by each eigenvector (finding the dot product).
PCA1 = np.dot(data, evecs[:, 0])
PCA2 = np.dot(data, evecs[:, 1])

# Plot the projection values
plt.figure(figsize=(10, 8))
plt.scatter(PCA1, PCA2, alpha=0.5)
plt.title('PCA Scatter Plot')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```
